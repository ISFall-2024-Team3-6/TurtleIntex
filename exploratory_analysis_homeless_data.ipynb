```python
import ssl

# Override the default SSL context
ssl._create_default_https_context = ssl._create_unverified_context

import pandas as pd

df = pd.read_csv('/Users/portergraham/Downloads/Simplified_PIT_data.csv')
df.head()
```

```python
# added a state column to the dataframe to allow for more statistical analysis

def extract_state(coc_number):
    return coc_number.split('-')[0]

df['State'] = df['CoC_Number'].apply(extract_state)
```

```python
df
```

```python
# Calculate the average count of unsheltered homeless people per state per year

average_unsheltered_per_state_year = df.groupby(['State', 'Year'])['UnshelteredTotal'].mean().reset_index()

print(average_unsheltered_per_state_year)
```

```python
# Calculate the average count of unsheltered homeless people per state in 2017

df_2017 = df[df['Year'] == 2017]

average_unsheltered_2017 = df_2017.groupby('State')['UnshelteredTotal'].mean().reset_index()

average_unsheltered_2017
```

```python
# Calculate the average count of unsheltered homeless people per state in 2022

df_2022 = df[df['Year'] == 2022]

# Calculate the average count of unsheltered homeless people per state for the year 2022
average_unsheltered_2022 = df_2022.groupby('State')['UnshelteredTotal'].mean().reset_index()

average_unsheltered_2022
```

```python
# Calculate the average count of unsheltered homeless people per year

average_unsheltered = df.groupby('Year')['UnshelteredTotal'].mean().reset_index()

average_unsheltered
```

```python
# Determine if there is a statistically significant difference between total unsheltered homeless people and the year 

from scipy import stats

groups = df['Year'].unique()  
group_labels = []               
for g in groups:                
  group_labels.append(df[df.Year == g]['UnshelteredTotal'])  
f, p =  stats.f_oneway(*group_labels) 

print('F: ' + str(round(f, 4)))
print('p: ' + str(round(p, 4)))



# at least one year's count of unsheltered homeless people are statistically significantly different from the rest of the other year's counts of unsheltered homeless people. We know this because the p-value is less than 0.05. 
```

```python
# tukey test to compare pairs of years with the count of unsheltered homeless people 

from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=df['UnshelteredTotal'],
                          groups=df['Year'],
                          alpha=0.05)

print(tukey)


# Only 2 pairs have an adjusted p-value of less than 0.05. In that case we can reject the null hypothesis and say 
# there is a statistically significant difference in count of unsheltered homeless people
# between years 2020-2021 and 2021-2022. 
# the year 2021 is the anomaly
```
